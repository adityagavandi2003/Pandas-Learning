**Lesson 4 â€” Basic Operations & Aggregations in Pandas**

---



```python
import pandas as pd

data = {
    "Product": ["Apples", "Bananas", "Oranges", "Apples", "Bananas"],
    "Quantity": [10, 5, 8, 7, 3],
    "Price": [50, 20, 30, 50, 20]
}

df = pd.DataFrame(data)
df["Sales"] = df["Quantity"] * df["Price"]
print(df)
```

1. Sorting

```python
df.sort_values("Sales", ascending=False)
```

2. Aggregations (single column)

```python
max_quantity = df['Quantity'].max() # find max quantity
print("Max Quantity",max_quantity)
min_quantity = df["Quantity"].min() # find min quantity
print("Min Quantity",min_quantity)
sum_quantity = df["Quantity"].sum() # find sum quantity
print("Sum Quantity",sum_quantity)
mean_quantity = df['Quantity'].mean() # find mean quantity
print("Mean",mean_quantity)
```

3. Multiple Aggregations

```python
df[["Quantity","Sales"]].agg(["sum","mean"])
```

4. GroupBy

```python
df.groupby("Product")["Sales"].sum()
```

Sort data in descending order

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
```

5. Value Counts

```python
df["Product"].value_counts()
```

6. Renaming Columns

```python
df.rename(columns={"Sales":"Total_Sales"})
```
