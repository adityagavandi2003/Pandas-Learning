```python
# Create a pd.Series called scores_s with names as the index.
name = ["Aditya","Arav","Janit","Raghav","Esha"]
scores = [72, 85, 90, 66, 79]
# create a series
scores_s = pd.Series(scores,index=name)
scores_s
```

```python
# What’s the average score? The highest and lowest?
scores_s.mean()
```

```python
# Who scored ≥ 80? Return a filtered Series, not just names.
high = scores_s[scores_s > 80]
high
```

```python
# Update Esha’s score to 75, then show the Series sorted descending.
update_Esha = scores_s["Esha"] = 75
scores_s.sort_values(ascending=False)
```

```python
# Bonus: Add “Rahul”: 88 to the Series (hint: assignment by new label) and re-compute the average.
scores_s["Rahul"] = 88
print(scores_s)
scores_s.mean()
```
