```python
# Filtering and Conditional Selection
import pandas as pd

data = {
    "Product": ["Apples", "Bananas", "Oranges", "Apples", "Bananas"],
    "Quantity": [10, 5, 8, 7, 3],
    "Price": [50, 20, 30, 50, 20]
}
df = pd.DataFrame(data)
df["Total"] = df["Quantity"] * df["Price"]
print(df)
```

```python
df[df["Quantity"]>5] # quantity is greater than 5
```

2 Multiple Conditions \
Use & (AND), | (OR), and ~ (NOT).\
Always wrap conditions in paranthesis

---



```python
# Quantity > 5 AND Price == 50
df[(df["Quantity"]>5) & (df["Price"]==50)]
```

```python
# Quantity < 6 OR Product == "Oranges"
df[(df["Quantity"]>5) | (df["Product"]== "Oranges")]
```

3. Filtering by Values (isin)

---

```python
df[df["Product"].isin(["Apples","Oranges"])]
```

4. Filtering by String Methods

---



```python
# Products starting with B → Bananas
df[df['Product'].str.startswith("B")]
```

5. Query Method (Optional shortcut)

---



```python
df.query("Quantity > 5 and Price == 50")
```

Mini Practice — Filtering

---



```python
# Show only rows where Quantity ≥ 8.
df[df["Quantity"] >= 8]
```

```python
# Get rows where Product = "Bananas".
df[df["Product"] == "Bananas"]
```

```python
# Get rows where Quantity > 5 AND Price < 40.
df[(df["Quantity"]> 5) & (df["Price"]< 40)]
```

```python
# Show rows where Product is either Apples or Oranges.
df[df["Product"].isin(["Apples","Oranges"])]
```

```python
# Show rows where Product does NOT start with 'B'.
df[~df['Product'].str.startswith('B')]
```
